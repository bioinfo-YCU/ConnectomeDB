```python
import requests
import pandas as pd
import numpy as np
import sys
import os 

sys.path.append(os.path.abspath("src"))  # Add src directory to path
from createDataTable import gene_pair0

# Get all unique genes
ligand_list = gene_pair0["Ligand"].tolist()
receptor_list = gene_pair0["Receptor"].tolist()
unique_genes = list(set(ligand_list + receptor_list))  # Combine and remove duplicates
len(unique_genes)

connectomeDB = pd.read_table("data/connectome_j.tsv", sep="\t")

#connectomeDB = connectomeDB[connectomeDB["Taxon"]== "Human"]
connectomeDB = connectomeDB.drop(columns=["Localization", "Taxon"] + [col for col in connectomeDB.columns if col.startswith("F5_")])
old_gene_symbols=set(connectomeDB['ApprovedSymbol'].unique())
intersection = pd.Series(list(old_gene_symbols.intersection(unique_genes)))
len(intersection)
len(old_gene_symbols)
```

```python
old_gene_symbols = list(old_gene_symbols)
#old_gene_symbols = old_gene_symbols[:100]
old_gene_symbols
```

```python
def get_ensembl_id(gene_symbol):
    url = f"https://rest.ensembl.org/xrefs/symbol/homo_sapiens/{gene_symbol}?content-type=application/json"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        if isinstance(data, list) and len(data) > 0 and 'id' in data[0]:
            return data[0]['id']  # Ensembl Gene ID
    return None  # Return None if not found

def get_latest_hgnc_symbol(ensembl_id):
    url = f"https://rest.ensembl.org/lookup/id/{ensembl_id}?content-type=application/json"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        if 'display_name' in data:
            return data['display_name']  # Latest HGNC Gene Symbol
    return None  # Return None if not found

def update_gene_symbols(old_symbols):
    results = []
    for symbol in old_symbols:
        ensembl_id = get_ensembl_id(symbol)
        new_symbol = get_latest_hgnc_symbol(ensembl_id) if ensembl_id else None
        results.append((symbol, ensembl_id, new_symbol))
    
    return pd.DataFrame(results, columns=['Old_Symbol', 'Ensembl_ID', 'Updated_Symbol'])

# Example usage
updated_symbols = update_gene_symbols(old_gene_symbols)
print(updated_symbols)
```
